### $\text{1. Theoretical framework}$
In the paraxial approximation, we propose a field of the form $U(\mathbf{r}) = \psi(\mathbf{r}) \mathrm{e}^{\mathrm{i} k z}$, where $\psi(\mathbf{r})$ is a function slowly varying along the $z$-axis. This assumption let us derive the paraxial wave equation
$$
\left( \nabla_{\perp}^2 + 2 \mathrm{i} k \dfrac{\partial}{\partial z} \right) \psi(\mathbf{r}) = 0,
$$

where $k$ is the wave number and $\nabla_{\perp}$ is the transveral laplacian:
$$
\nabla_{\perp} \equiv \dfrac{\partial^2}{\partial x^2} + \dfrac{\partial^2}{\partial y^2}.
$$

As a reminder, the wave number is related to the wavelength through the equation $k = 2 \pi / \lambda$. We also recall that the Fresnel approximation is represented through
$$
\psi(x, y, z) = \dfrac{1}{\mathrm{i} \lambda z} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \mathrm{d}x_{0} \, \mathrm{d}y_{0}.
$$

To prove that the Fresnel approximation is a solution to the paraxial equation, we compute the derivatives:
$$
\begin{align*}
\dfrac{\partial}{\partial x} \psi(\mathbf{r})
&= \dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) (x - x_{0}) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace  \, \mathrm{d}x_{0} \, \mathrm{d}y_{0},\\
%
\dfrac{\partial}{\partial y} \psi(\mathbf{r})
&= \dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) (y - y_{0}) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \, \mathrm{d}x_{0} \, \mathrm{d}y_{0}, \\
%
\dfrac{\partial^2}{\partial x^2} \psi(\mathbf{r})
&= \dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \left[ 1 + \dfrac{\mathrm{i} (x - x_{0})^2}{z} \right] \, \mathrm{d}x_{0} \, \mathrm{d}y_{0}, \\
%
\dfrac{\partial^2}{\partial y^2} \psi(\mathbf{r})
&= \dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \left[ 1 + \dfrac{\mathrm{i} (y - y_{0})^2}{z} \right] \, \mathrm{d}x_{0} \, \mathrm{d}y_{0}.
\end{align*} $$

With this, we notice that
$$
\nabla_{\perp}^2 \psi(\mathbf{r}) = \dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{z} + 2 \right\rbrace.
$$

The next step is obtaining the $z$ derivative of $\psi(\mathbf{r})$:
$$
\begin{align*}
2 \mathrm{i} k \dfrac{\partial}{\partial z} \psi(\mathbf{r}) 
&= -\dfrac{2 k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} + 1 \right\rbrace \\
&= -\dfrac{k}{\lambda z^2} \iint \psi(x_{0}, y_{0}, 0) \exp\left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{2 z} \right\rbrace \left\lbrace \dfrac{\mathrm{i} k [(x - x_{0})^2 + (y - y_{0})^2]}{z} + 2 \right\rbrace.
\end{align*}
$$

Evidently, the Fresnel approximation satisfies the paraxial equation. On the other hand, we could also examine circularly symmetric solutions to the paraxial equation by writing it down in cylindrical coordinates:
$$
\left[ \dfrac{1}{r} \dfrac{\partial}{\partial r} \left( r \dfrac{\partial}{\partial r} \right) + 2 \mathrm{i} k \dfrac{\partial}{\partial z} \right] \Psi(\mathbf{r}) = 0.
$$

Here, we propose a solution with the following structure:
$$
\Psi(\mathbf{r}) = \exp\left\lbrace \mathrm{i} \left[ p(z) + \dfrac{k r^2}{2 R(z)} \right] \right\rbrace.
$$

By introducing a characteristic length (called Gaussian waist) and defined as
$$
w_{0}^2 = \dfrac{2 z_{\mathrm{R}}}{k},
$$

we could prove that the paraxial equation solution can be expressed as
$$
\Psi(\mathbf{r}) = \dfrac{w_{0}}{w(z)} \exp\left[ -\dfrac{r^2}{w^2(z)} \right] \exp\left\lbrace \mathrm{i} \left[ k z - \arctan\left( \dfrac{z}{z_{\mathrm{R}}} \right) \right] \right\rbrace \exp\left[ \mathrm{i} \dfrac{k r^2}{2 R(z)} \right].
$$

Here, $\Psi(\mathbf{r})$ represents a Gaussian beam. From this equation, we define the following expressions:
- $w(z)$ is the radius at which the field amplitude falls to $1/e$, at the plane $z$ along the beam.
- $w_{0} = w(0)$ is the waist radius.
- $R(z)$ is the wavefront radius of curvature.
- $z_{\mathrm{R}}$ is the Rayleigh range.
- $\varphi(z) = \arctan(z / z_{\mathrm{R}})$ is the Gouy phase.

Other relevant relations are the following:
$$
\begin{align*}
    w^2(z) &= w_{0}^2 \left[ 1 + \left( \dfrac{z}{z_{\mathrm{R}}} \right)^2 \right], \\
    %
    R(z)  &= z \left[ 1 + \left( \dfrac{z_{\mathrm{R}}}{z} \right)^2 \right].
\end{align*}
$$

To obtain the analytic expression of a beam propagating through an $\mathrm{ABCD}$ medium, we first remember we can characterize some optical elements with a $2 \times 2$ matrix:
$$
\begin{bmatrix}
A & B \\ C & D
\end{bmatrix}.
$$

By the $\mathrm{ABCD}$ law for Gaussian beams, the radius of curvature changes from $R_{1}$ to $R_{2}$ in such a way that
$$
R_{2} = \dfrac{A R_{1} + B}{C R_{1} + D},
$$

with its inverse given by
$$
\dfrac{1}{R_{2}} = \dfrac{C + \frac{D}{R_{1}}}{A + \frac{B}{R_{1}}}.
$$

To illustrate this, let us consider a generalized Helmholtz-Gauss ($\text{HzG}$) beam, which are approximations of non-diffractive beams, and study its propagation through an $\mathrm{ABCD}$ medium. The transversal distribution of the $\text{HzG}$ in a plane $z = z_{1}$ is the result of the product between a Gaussian factor and a scalar function, $W$, which is a solution to the non-diffractive Helmholtz equation:
$$
\mathrm{HzG}_{1}(\mathbf{r}_{1}) = \exp\left( \dfrac{\mathrm{i} k r_{1}^2}{2 R_{1}} \right) W(\mathbf{r}_{1}; \kappa_{1}),
$$

where $R_{1}$, in general, is a complex number satisfying the condition $R_{1} \in \mathrm{Im}\lbrace R_{1} \rbrace <  0$; $\kappa$ is the wave vector transveral component. The function $W(\mathbf{r}_{1}; \kappa_{1})$ has the following form:
$$
W(\mathbf{r}_{1} ; \kappa_{1}) = \int\limits_{-\pi}^{\pi} g(\varphi) \exp \lbrace \mathrm{i} \kappa_{1} [ x_{1} \cos(\varphi) + y_{1} \sin(\varphi) ] \rbrace \: \mathrm{d}\varphi.
$$

We can employ the generalized diffraction integral to explain the propagation from a plane $z_{1}$ to a plane $z_{2}$ through a $\mathrm{ABCD}$ medium,
$$
\mathrm{HzG}_{2}(\mathbf{r}_{2}) = \dfrac{k \exp(\mathrm{i} k L)}{2 \pi \mathrm{i} B} \iint\limits_{-\infty}^{\infty} \mathrm{HzG}_{1}(\mathbf{r}_{1}) \exp \left\lbrace \dfrac{\mathrm{i} k}{2 B} [ A r_{1}^2 - 2 (x_{2} x_{1} + y_{2} y_{1}) + D r_{2}^2 ]  \right\rbrace \: \mathrm{d}x_{1} \: \mathrm{d}y_{1},
$$

where $\mathrm{r}_{2} = (x_{2}, y_{2})$ and $L$ is the optical path length from plane $z_{1}$ to plane $z_{2}$, measured from the optical axis. Making use of these definitions, we first notice that
$$
\mathrm{HzG}_{2}(\mathbf{r}_{2}) = \dfrac{k \mathrm{e}^{\mathrm{i} k L}}{2 B \pi \mathrm{i}} \int\limits_{-\infty}^{\infty} \int\limits_{-\infty}^{\infty} \mathrm{e}^{\frac{\mathrm{i} k r_{1}^2}{2 R_{1}}} \int\limits_{-\pi}^{\pi} g(\varphi) \mathrm{e}^{\mathrm{i} \kappa_{1} \left[ x_{1} \cos(\varphi) + y_{1} \sin(\varphi) \right]} \: \mathrm{d}\varphi \: \mathrm{e}^{\frac{\mathrm{i} k \left[ A R_{1}^2 - 2 (x_{1} x_{2} + y_{1} y_{2} + D r_{2}^2) \right]}{2 B}} \: \mathrm{d} x_{1} \: \mathrm{d}y_{1}.
$$

Since the limits are constant, we are able to use Fubini's theorem and reexpress the double integral as the product of two individual integrals:
$$
\mathrm{HzG}_{2}(\mathbf{r}_{2}) = \dfrac{k \mathrm{e}^{\mathrm{i} k L}}{2 B \pi \mathrm{i}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}} \int\limits_{-\pi}^{\pi} \left[ \int\limits_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i} \left\lbrace \frac{k x_{1}^2}{2 R_{1}} + \kappa_{1} x_{1} \cos(\varphi) + \frac{k [A x_{1}^2 - 2 x_{1} x_{2}]}{2 B} \right\rbrace} \:\mathrm{d}x_{1} \right] \left[ \int\limits_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i} \left\lbrace \frac{k y_{1}^2}{2 R_{1}} + \kappa_{1} y_{1} \sin(\varphi) + \frac{k [A y_{1}^2 - 2 y_{1} y_{2} ]}{2 B} \right\rbrace} \: \mathrm{d}y_{1} \right] \:\mathrm{d}\varphi.
$$

We have taken out the elements independent of $\mathbf{r}_{1}$. We realize that the resulting integrals have the same shape as the Fresnel integrals. The only thing we need to do is completing the square and using a substitution to write them in the general form:
$$
\int\limits_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i} x^2} \: \mathrm{d}x = \int\limits_{-\infty}^{\infty} \cos(x^2) \: \mathrm{d}x + \mathrm{i} \int\limits_{-\infty}^{\infty} \sin(x^2) \: \mathrm{d}x = \dfrac{\sqrt{2 \pi}}{2} + \dfrac{\sqrt{2 \pi}}{2} \mathrm{i}.
$$

For time's sake, we can evaluate these integrals using $\texttt{Mathematica}$, yielding the following results:
$$
\begin{align*}
    \int\limits_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i} \left\lbrace \frac{k x_{1}^2}{2 R_{1}} + \kappa_{1} x_{1} \cos(\varphi) + \frac{k [ A x_{1}^2 - 2 x_{1} x_{2} ]}{2 B} \right\rbrace} \: \mathrm{d} x_{1}
    &= \dfrac{\mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[k x_{2} - B \cos(\varphi) \kappa_{1} \right]^2}{2 B k(A R_{1} + B)}} \sqrt{2 \pi}}{\sqrt{-\frac{\mathrm{i} k (A R_{1} + B)}{B R_{1}}}}, \\
    \int\limits_{-\infty}^{\infty} \mathrm{e}^{\mathrm{i} \left\lbrace \frac{k y_{1}^2}{2 R_{1}} + \kappa_{1} y_{1} \sin(\varphi) + \frac{k [ A y_{1}^2 - 2 y_{1} y_{2}]}{2 B} \right\rbrace} \: \mathrm{d}y_{1}
    &= \dfrac{\mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[k y_{2} - B \sin(\varphi) \kappa_{1} \right]^2}{2 B k(A R_{1} + B)}} \sqrt{2\pi}}{\sqrt{-\frac{\mathrm{i} k (A R_{1} + B)}{B R_{1}}}}.
\end{align*}
$$

Combining the results, we get
$$
\begin{align*}
    \mathrm{HzG}_{2}(\mathbf{r}_{2}) 
    &= \dfrac{k\mathrm{e}^{\mathrm{i} k L}}{2 B \pi \mathrm{i}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}}\int\limits_{-\pi}^{\pi} g(\varphi) \left[\dfrac{\mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[ k x_{2} - B \cos(\varphi) \kappa_{1} \right]^2}{2 B k(A R_{1} + B)}} \sqrt{2 \pi}}{\sqrt{-\frac{\mathrm{i} k ( A R_{1} + B )}{B R_{1}}}} \right] \left[\dfrac{\mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[k y_{2} - B \sin(\varphi) \kappa_{1} \right]^2}{2 B k( A R_{1} + B )}} \sqrt{2 \pi}}{\sqrt{-\frac{\mathrm{i} k (A R_{1} + B)}{B R_{1}}}}\right] \: \mathrm{d}\varphi, \\
    %
    &= \dfrac{k\mathrm{e}^{\mathrm{i} k L}}{2 B \pi \mathrm{i}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}} \dfrac{2 \pi}{\frac{\mathrm{i} k (A R_{1} + B)}{B R_{1}}} \int\limits_{-\pi}^{\pi} g(\varphi) \mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[ k^2 x_{2}^2 - 2 B k \kappa_{1} \cos(\varphi) x_{2} + B^2 \cos^2(\varphi) \kappa_{1}^2 + k^2 y_{2}^2 - 2 B k \kappa_{1} \sin(\varphi) y_{2} + B^2 \sin^2(\varphi) \kappa_{1}^2 \right]}{2 B k(A R_{1} + B)}}\: \mathrm{d} \varphi,\\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}} \int\limits_{-\pi}^{\pi} g(\varphi) \mathrm{e}^{-\frac{\mathrm{i} R_{1} \left[ k^2 x_{2}^2 - 2 B k \kappa_{1} \cos(\varphi) x_{2} + B^2 \cos^2(\varphi) \kappa_{1}^2 + k^2 y_{2}^2 - 2 B k \kappa_{1} \sin(\varphi) y_{2} + B^2 \sin^2(\varphi) \kappa_{1}^2 \right]}{2 B k(A R_{1} + B)}} \: \mathrm{d} \varphi,\\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}} \mathrm{e}^{-\frac{\mathrm{i} R_{1} k^2(x_{2}^2 + y_{2}^2)}{2 B k (A R_{1} + B)}} \mathrm{e}^{-\frac{\mathrm{i} R_{1} B^2 \kappa_{1}^2}{2 B k(A R_{1} + B)}} \int\limits_{-\pi}^{\pi} g(\varphi) \mathrm{e}^{\frac{\mathrm{i} R_{1}(2 B k) \kappa_{1} \left[ \cos(\varphi) x_{2} + \sin(\varphi)y_{2} \right]}{(2 B k)(A R_{1} + B)}} \: \mathrm{d} \varphi, \\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2 B}} \mathrm{e}^{\frac{-\mathrm{i} k r_{2}^2 R_{1}}{2 B (A R_{1} + B)}} \mathrm{e}^{\frac{B \kappa_{1}}{2 k \mathrm{i}} \frac{\kappa_{1}}{A + \frac{B}{R_{1}}}} \underbrace{\int\limits_{-\pi}^{\pi} g(\varphi) \mathrm{e}^{\mathrm{i} \kappa_{2} \left[ \cos(\varphi) x_{2} + \sin(\varphi) y_{2}\right]} \: \mathrm{d} \varphi}_{W(\mathbf{r}_{2}; \kappa_{2})}, \\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k D r_{2}^2}{2B}} \mathrm{e}^{\frac{-\mathrm{i} k r_{2}^2 R_{1}}{2 B (A R_{1} + B)}} \mathrm{e}^{\frac{B \kappa_{1} \kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}), \\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} kL}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k r_{2}^2}{2 B} \left(D - \frac{R_{1}}{A R_{1} + B}\right)} \mathrm{e}^{\frac{B \kappa_{1} \kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}), \\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k r_{2}^2}{2 B} \left(\frac{A D R_{1} + B D - R_{1}}{A R_{1} + B}\right)} \mathrm{e}^{\frac{B \kappa_{1}\kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}).
\end{align*}
$$

Now, the determinant of the $\mathrm{ABCD}$ matrix should be equal to $1$, with implies that $\mathrm{AD} - \mathrm{BC} = 1$. Isolating for $\mathrm{AD}$, we get
$$
\begin{align*}
    \mathrm{HzG}_{2}(\mathbf{r}_{2}) 
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k r_{2}^2}{2B} \left(\frac{B C R_{1} + R_{1} + B D - R_{1}}{A R_{1} + B}\right)} \mathrm{e}^{\frac{B \kappa_{1} \kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}), \\
    %
    &= \dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k r_{2}^2}{2} \left(\frac{C R_{1} + D}{A R_{1} + B}\right)} \mathrm{e}^{\frac{B \kappa_{1} \kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}), \\
    %
    &= \underbrace{\dfrac{\mathrm{e}^{\mathrm{i} k L}}{A + \frac{B}{R_{1}}} \mathrm{e}^{\frac{\mathrm{i} k r_{2}}{2 R_{2}}}}_{\text{GB}(\mathbf{r}_{2}, R_{2})} \mathrm{e}^{\frac{B \kappa_{1} \kappa_{2}}{2 k \mathrm{i}}} W(\mathbf{r}_{2}; \kappa_{2}), \\
    %
    &= \mathrm{e}^{\frac{\kappa_{1} \kappa_{2} B}{2 k \mathrm{i}}} \text{GB}(\mathbf{r}_{2}, R_{2})W(\mathbf{r}_{2}; \kappa_{2}).
\end{align*}
$$

To start with, we will study the behavior of a Gaussian beam propagating through (a) a free-space segment of $1.5 z_{\mathrm{R}}$ and (b) in a same length segment with the beam focused with a thin, convergent lens with focal distance $f = 5 \: \mathrm{cm}$. We first declare the physical and numerical parameters:

```python
def gauss_propagation(wavelength, grid_points, focal_length = None):
    # Physical parameters
    w0 = 5e-4                       # Beam waist radius [m]
    k = 2 * np.pi / wavelength      # Wavenumber
    zR = (w0 ** 2 * k) / 2          # Rayleigh range
    z_max = 1.5 * zR                # Maximum propagation distance

    # Numerical grid
    N = grid_points                 # Number of grid points
    window_size = 3 * w0            # Window size
    dx = 2 * window_size / N        # Spatial step
    max_spatial_freq = np.pi / dx   # Maximum spatial frequency

    # Coordinates
    grid_normalized = np.arange(-N / 2, N / 2) * (2 / N)
    x = grid_normalized * window_size
    y = grid_normalized * window_size
    z = np.linspace(0, z_max, N)

    # Frequency domain
    kx = grid_normalized * max_spatial_freq
    ky = grid_normalized * max_spatial_freq

    # Meshgrids
    X, Y = np.meshgrid(x, y)
    Kx, Ky = np.meshgrid(kx, ky)

    # Initial field (Gauss)
    r = X ** 2 + Y ** 2
    GB0 = np.exp(-r / w0 ** 2)

    # Propagation
    GB0_fft = np.fft.fftshift(np.fft.fft2(GB0))
    GB_prop = np.zeros((N, len(z)), dtype = complex)

    for n, zn in enumerate(z):
        propagator = np.exp(-1j * zn / (2 * k) * (Kx ** 2 + Ky ** 2))
        GBz = np.fft.ifft2(GB0_fft * propagator) * np.exp(1j * k * zn)
        GB_prop[:, n] = GBz[N // 2, :]

    Z_mesh, Y_mesh = np.meshgrid(z, y)
    intensity = np.abs(GB_prop) ** 2

    if focal_length is None:
        return X, Y, GB0, Z_mesh, Y_mesh, intensity
    else:
        f = focal_length * zR
        thin_lens = np.exp(-1j * k * r / (2 * f))

        # Initial field (with lens)
        GB0_lens = GB0 * thin_lens
        
        # Propagation
        GB0_lens_fft = np.fft.fftshift(np.fft.fft2(GB0_lens))
        GB_lens_prop = np.zeros((N, len(z)), dtype = complex)

        for n, zn in enumerate(z):
            propagator = np.exp(-1j * zn / (2 * k) * (Kx ** 2 + Ky ** 2))
            GBz = np.fft.ifft2(GB0_lens_fft * propagator) * np.exp(1j * k * zn)
            GB_lens_prop[:, n] = GBz[N // 2, :]
        intensity_GB_lens = np.abs(GB_lens_prop) ** 2

        return X, Y, GB0, Z_mesh, Y_mesh, intensity, GB0_lens, intensity_GB_lens

def plot_initial_condition(X, Y, field, title):
    plt.figure(figsize = (16, 9))
    plt.pcolormesh(X, Y, np.abs(field), shading = 'auto', cmap = 'jet')
    plt.xlabel(r"$x \: [\mathrm{m}]$")
    plt.ylabel(r"$y \: [\mathrm{m}]$")
    plt.ticklabel_format(style = 'scientific', axis = 'both', scilimits = (0, 0))
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()

def plot_propagation(Z, Y, intensity, title):
    plt.figure(figsize = (16, 9))
    plt.pcolormesh(Z, Y, intensity, shading = 'auto', cmap = 'jet')
    plt.xlabel(r"$z \: [\mathrm{m}]$")
    plt.ylabel(r"$y \: [\mathrm{m}]$")
    plt.ticklabel_format(style = 'scientific', axis = 'both', scilimits = (0, 0))
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()
```

```python
X, Y, GB0, Z_mesh, Y_mesh, intensity, GB0_lens, intensity_GB_lens = gauss_propagation(500e-9, 2 ** 8, 0.5)

plot_initial_condition(
    X, Y, GB0,
    rf'Initial Gaussian Beam'
    )

plot_propagation(
    Z_mesh, Y_mesh, intensity,
    rf'Gaussian Beam free-space propagation'
    )

plot_initial_condition(
    X, Y, GB0_lens,
    rf'Initial Gaussian Beam through a thin lens'
    )

plot_propagation(
    Z_mesh, Y_mesh, intensity_GB_lens,
    rf'Gaussian Beam propagation through a thin lens ($f = 0.5 z_{{\mathrm{{R}}}}$)'
    )
```

Now, we will study the behavior of a Bessel-Gauss beam ($\mathrm{BGB}$), but instead of focusing the beam, we will use different beam waist values.

```python
def bessel_gauss_propagation(w0, wavelength = 632.8e-9, m = 0, N = 2 ** 8, z_factor = 1.5):
    # Physical parameters
    k = 2 * np.pi / wavelength          # Wavenumber
    kt = 8665                           # Transveral wavenumber
    zR = (w0 ** 2 * k) / 2              # Rayleigh range
    z_max = z_factor * zR               # Propagation distance
    
    # Numerical grid
    window_size = 2 * w0                # Simulation window
    dx = 2 * window_size / N            # Spatial step
    max_spatial_freq = np.pi / dx       # Max spatial frequency
    
    # Coordinates
    grid_normalized = np.arange(-N / 2, N / 2) * (2 / N)
    x = grid_normalized * window_size
    y = grid_normalized * window_size
    z = np.linspace(0, z_max, N)
    
    # Frequency domain
    kx = grid_normalized * max_spatial_freq
    ky = grid_normalized * max_spatial_freq

    # Meshgrids
    X, Y = np.meshgrid(x, y)
    Kx, Ky = np.meshgrid(kx, ky)
    
    # Initial field (Bessel-Gauss)
    r = np.sqrt(X ** 2 + Y ** 2)
    Bessel = sc.jv(m, kt * r)
    BGB0 = np.exp(-r ** 2 / w0 ** 2) * Bessel * np.exp(1j * m * np.arctan2(Y, X))
    
    # Propagation
    BGB0_fft = np.fft.fftshift(np.fft.fft2(BGB0))
    BGB_prop = np.zeros((N, len(z)), dtype = complex)
    
    for n, zn in enumerate(z):
        propagator = np.exp(-1j * zn / (2 * k) * (Kx**2 + Ky**2))
        BGBz = np.fft.ifft2(BGB0_fft * propagator) * np.exp(1j * k * zn)
        BGB_prop[:, n] = BGBz[N // 2, :]
    
    # Prepare for plotting
    Z_mesh, Y_mesh = np.meshgrid(z, y)
    intensity = np.abs(BGB_prop) ** 2
    
    return X, Y, BGB0, Z_mesh, Y_mesh, intensity
```

```python
# Waist values
waists = [0.5e-3, 1.5e-3, 3.0e-3]  # Beam waists to test

for w0 in waists:
    X, Y, BGB0, Z_mesh, Y_mesh, intensity = bessel_gauss_propagation(w0)
    
    plot_initial_condition(
        X, Y, BGB0,
        rf'Initial Bessel-Gauss Beam ($w_{0} = {w0 * 1e3:.1f} \: \mathrm{{mm}}$)'
    )
    
    plot_propagation(
        Z_mesh, Y_mesh, intensity,
        rf'Bessel-Gauss Beam propagation ($w_{0} = {w0 * 1e3:.1f} \: \mathrm{{mm}}$)'
    )
```
