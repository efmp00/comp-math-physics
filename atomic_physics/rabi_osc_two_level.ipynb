### $\text{1. Introduction}$
##### $\textit{1.1: Hamiltonian of an atom in a two-level system}$
Suppose we have an atom in a two-level system, *i.e.* it can only access two eigenstates, $|1\rangle$ and $|2\rangle$. We can express the Hamiltonian as
$$
\hat{H} = \hat{H}_{0} + \hat{H}_{\mathrm{atom}}(t),
$$

where $\hat{H}_{\mathrm{atom}}(t)$ describes the atom-interaction and perturbates the eigenfunctions of $\hat{H}_{0}$ (Foot, 2003). The (unperturbed) Hamiltonian $\hat{H}_{0}$ produces the following energies:
$$
\begin{align*}
    \hat{H}_{0} | 1 \rangle &= \hbar \omega_{1} | 1 \rangle, \\ 
    \hat{H}_{1} | 2 \rangle &= \hbar \omega_{2} | 2 \rangle.
\end{align*}
$$

On the other hand, the atom-light interaction Hamiltonian considers an oscillating electric field of a plane monochromatic wave,
$$
\hat{H}_{\mathrm{atom}}(t) = -\mathbf{d} \bullet \mathbf{E},
$$

where the electric field itself is given by
$$
\mathbf{E} = 2 E_{0} \cos(\mathbf{k} \bullet \mathbf{r} - \omega t) \hat{\mathbf{e}}
$$

and the electric dipole by $\mathbf{d} = -e \mathbf{r}$. As usual, $\mathbf{r}$ is the position vector, $\mathbf{k}$ is the wave vector, $\omega$ is the angular frequency, and $\hat{\mathbf{e}}$ is the polarization. 

##### $\textit{1.2: Solution of the Schrödinger equation}$
As expected, the system must follow Schrödinger's equation:
$$
\mathrm{i} \hbar \dfrac{\partial}{\partial t} \Psi(t) = [\hat{H}_{0} + \hat{H}_{\mathrm{atom}(t)}] \Psi(t).
$$

We propose a solution where the probability is time-dependent with the following form:
$$
\Psi(t) = c_{1}(t) | 1 \rangle + c_{2} (t) | 2 \rangle.
$$

Thus we insert our proposed solution into Schrödinger's equation. We start with the left-hand side:
$$
\begin{align*}
    [\hat{H}_{0} + \hat{H}_{\mathrm{atom}}(t)] \Psi(t)
    &= \hat{H}_{0} [ c_{1}(t) | 1 \rangle + c_{2} (t) | 2 \rangle ] + \hat{H}_{\mathrm{atom}}(t) [ c_{1}(t) | 1 \rangle + c_{2} (t) | 2 \rangle ] \\
    &= \hbar \omega_{1} c_{1}(t) | 1 \rangle + \hbar \omega_{2} c_{2}(t) | 2 \rangle + \hat{H}_{\mathrm{atom}}(t) c_{1}(t) | 1 \rangle + \hat{H}_{\mathrm{atom}}(t) c_{2}(t) | 2 \rangle.
\end{align*}
$$

Putting everything together, we see that
$$
\mathrm{i} \dfrac{\partial}{\partial t} c_{1}(t) | 1 \rangle + \mathrm{i} \dfrac{\partial}{\partial t} c_{2}(t) | 2 \rangle = c_{1}(t) [ \hbar \omega_{1} + \hat{H}_{\mathrm{atom}}(t) ] | 1 \rangle + c_{2}(t) [ \hbar \omega_{2} + \hat{H}_{\mathrm{atom}}(t) ] | 2 \rangle.
$$

At this point, there are two scenarios:
- Applying $| 1 \rangle$ yields
$$
\mathrm{i} \hbar \dfrac{\partial}{\partial t} c_{1}(t) = \hbar \omega_{1} c_{1} (t) + \hat{H}_{\mathrm{atom}}^{12} c_{2}(t).
$$
- Applying $| 2 \rangle$ yields
$$
\mathrm{i} \hbar \dfrac{\partial}{\partial t} c_{2}(t) = \hbar \omega_{2} c_{2} (t) + \hat{H}_{\mathrm{atom}}^{21} c_{1}(t).
$$

Notice that both equations are dependent on a term of the form $H_{\mathrm{atom}}^{i j} = \langle i | \hat{H}_{\mathrm{atom}} | j \rangle$. This would be equivalent to writing
$$
\begin{align*}
    H_{\mathrm{atom}}^{i j} 
    &= -\langle i | \mathbf{d} \bullet \mathbf{E} | j \rangle \\
    &= -\langle i | \mathbf{d} \bullet \hat{\mathbf{e}} | j \rangle \cdot E_{0} \left[ \mathrm{e}^{\mathrm{i}(\mathbf{k} \bullet \mathbf{r} - \omega t)} + \mathrm{e}^{-\mathrm{i}(\mathbf{k} \bullet \mathbf{r} - \omega t)} \right].
\end{align*}
$$

##### $\textit{1.3: Dipole approximation}$
To simplify this last equation, we could neglect the phase $\mathbf{k} \bullet \mathbf{r}$ if we assume that the impinging electromagnetic wave has a wavelength bigger than the atom itself (Foot, 2003). Hence the last equation could be expressed as
$$
H_{\mathrm{atom}}^{i j} = -\langle i | \mathbf{d} \bullet \hat{\mathbf{e}} | j \rangle \cdot E_{0} \left( \mathrm{e}^{\mathrm{i} \omega t} + \mathrm{e}^{-\mathrm{i} \omega t} \right).
$$

This assumption is called **dipole approximation**. Now, we are left with a system of two Partial Differential Equations (PDE). To solve it, we introduce some substitutions:
$$
\begin{align*}
    c_{1}(t) &= \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t}, \\
    c_{2}(t) &= \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t}.
\end{align*}
$$

We start with $c_{1}(t)$. The left-hand side of the PDE for $c_{1}$ becomes
$$
\begin{align*}
    \mathrm{i} \hbar \dfrac{\partial}{\partial t} c_{1}(t)
    &= \mathrm{i} \hbar \dfrac{\partial}{\partial t} [ \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t} ] \\
    &= \mathrm{i} \hbar \dfrac{\partial}{\partial t} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t} + \hbar \omega_{1} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t}.
\end{align*}
$$

Now, the right-hand side:
$$
\hbar \omega_{1} c_{1}(t) + H_{\mathrm{atom}}^{12} c_{2}(t) = \hbar \omega_{1} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1}(t)} + H_{\mathrm{atom}}^{12} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t}.
$$

Equating both sides, we have
$$
\begin{align*}
    \mathrm{i} \hbar \dfrac{\partial \tilde{c}_{1}(t)}{\partial t} \mathrm{e}^{-\mathrm{i} \omega_{1} t} + \hbar \omega_{1} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t}
    &= \hbar \omega_{1} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t} + H_{\mathrm{atom}}^{12} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t}\\
    %
    \mathrm{i} \hbar \dfrac{\partial \tilde{c}_{1}(t)}{\partial t} \mathrm{e}^{-\mathrm{i} \omega_{1} t} 
    &= H_{\mathrm{atom}}^{12} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t} = -E_{0} \left( \mathrm{e}^{\mathrm{i} \omega t} + \mathrm{e}^{-\mathrm{i} \omega t} \right)\langle 1| \hat{\mathbf{d}} \bullet \hat{\mathbf{e}}| 2 \rangle \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t}.
\end{align*}
$$

We do the same for the second equation:
$$
\begin{align*}
    \mathrm{i} \hbar \dfrac{\partial c_{2}(t)}{\partial t} 
    &= \mathrm{i} \hbar \dfrac{\partial}{\partial t} \left[ \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t} \right] \\
    &= \mathrm{i} \hbar \dfrac{\partial}{\partial t} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t} + \hbar \omega_{2} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t}.
\end{align*}
$$

For the right-hand side, we have
$$
\begin{align*}
    \hbar \omega_{2} c_{2}(t) + H_{\mathrm{atom}}^{21} c_{1}(t) = \hbar \omega_{2} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t} + H_{\mathrm{atom}}^{21} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t}.
\end{align*}
$$

Equating both sides, we get
$$
\begin{align*}
    \mathrm{i} \hbar \dfrac{\partial \tilde{c}_{2}(t)}{\partial t} \mathrm{e}^{-\mathrm{i} \omega_{2} t} + \hbar \omega_{2} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2}t}
    &= \hbar \omega_{2} \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} \omega_{2} t} + H_{\mathrm{atom}}^{21} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1}t} \\
    \mathrm{i} \hbar \dfrac{\partial \tilde{c}_{2}(t)}{\partial t} \mathrm{e}^{-\mathrm{i} \omega_{2} t} 
    &= H_{\mathrm{atom}}^{21} \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t} = -E_{0} \left(\mathrm{e}^{\mathrm{i} \omega t} + \mathrm{e}^{-\mathrm{i} \omega t} \right) \langle 2 | \hat{\mathbf{d}} \bullet \hat{\mathbf{e}} | 1 \rangle \tilde{c}_{1}(t) \mathrm{e}^{-\mathrm{i} \omega_{1} t}.
\end{align*}
$$

In summary, we have the following result:
$$
\begin{align*}
    \dfrac{\partial \tilde{c}_{1}(t)}{\partial t} 
    &= \mathrm{i} \dfrac{d_{12} E_{0}}{\hbar} \left(\mathrm{e}^{\mathrm{i} \omega t} + \mathrm{e}^{-\mathrm{i} \omega t} \right) \tilde{c}_{2}(t) \mathrm{e}^{-\mathrm{i} (\omega_{2} - \omega_{1}) t}, \\
    \dfrac{\partial \tilde{c}_{2}(t)}{\partial t} 
    &= \mathrm{i} \dfrac{d_{21} E_{0}}{\hbar} \left(\mathrm{e}^{\mathrm{i} \omega t} + \mathrm{e}^{-\mathrm{i} \omega t}\right) \tilde{c}_{1}(t) \mathrm{e}^{\mathrm{i} (\omega_{2} - \omega_{1})t}.
\end{align*}
$$

From the PDF for $\tilde{c}_{1}(t)$, we define the detuning and Rabi frequency, respectively, as
$$
\begin{align*}
    \Delta &= \omega - \omega_{12}, \\
    \Omega &= -\dfrac{d_{12} E_{0}}{\hbar} ,
\end{align*}
$$

where $\omega_{12} = \omega_{2} - \omega_{1}$ and $d_{ij} = \langle i\,|\,\hat{\mathbf{d}} \bullet \hat{\mathbf{e}} \, | \, j \rangle$. With these expressions, the last PDE take the following shape:
$$
\begin{align*}
    \dfrac{\partial \tilde{c}_{1}(t)}{\partial t} &= -\mathrm{i} \Omega \mathrm{e}^{\mathrm{i} \Delta t} \tilde{c}_{2}(t), \\
    \dfrac{\partial \tilde{c}_{2}(t)}{\partial t} &= -\mathrm{i} \Omega^{*} \mathrm{e}^{-\mathrm{i} \Delta t} \tilde{c}_{1}(t).
\end{align*}
$$

It is relevant to mention that we ignore the terms that oscillate with a frequency of $\omega + \omega_{12}$. The physical reason for doing so is that we want to neglect fast oscillations since these will tend to go to zero; this procedure is called the **rotating wave approximation**. To solve this system of equations, we start from the ground state, assuming that $\tilde{c}_{1}(0) = 1$ and $\tilde{c}_{2}(0) = 0$. 

In addition, we take $\Delta$ and $\Omega$ as real-values. We will solve this system numerically. This system of equations could be solved analytically as well. Take into account that second last equation can be derived with respect to time and then use the last equation in it. From there, the remaining procedure is to deal with the algebra. We should arrive to the next solutions:
$$
\begin{align*}
    \tilde{c}_{1}(t) 
    &= \left[ \cos \left( \tilde{\Omega} t \right) - \dfrac{\mathrm{i} \Delta}{2 \tilde{\Omega}} \sin \left(\tilde{\Omega} t \right) \right] \mathrm{e}^{\frac{\mathrm{i} \Delta t}{2}}, \\
    %
    \tilde{c}_{2}(t) 
    &= -\dfrac{\mathrm{i} \Omega^{*}}{\tilde{\Omega}} \sin ( \tilde{\Omega} ) \mathrm{e}^{-\frac{\mathrm{i} \Delta t}{2}},
\end{align*}
$$

where
$$
    \tilde{\Omega} \equiv \sqrt{|\Omega|^2 + \dfrac{\Delta^2}{4}}.
$$

### $\text{2. Code}$
We start calling the libraries we will need for the numerical solution:

```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors

# Plot parameters
plt.rcParams.update({
    'lines.linewidth': 2,          # linewidth
    'text.usetex': True,           # LaTeX font
    'font.family': 'sans serif',   # Serif family
    'font.size': 20,               # font size
    'axes.titlesize': 26,          # title size
    'axes.grid': True,             # grid
    'grid.linestyle': "-.",        # grid style
    'axes.facecolor': '#ECECEC',   # Background color for the axes
    'figure.facecolor': '#FFFFFF', # Background color for the figure
    'legend.facecolor': '#FFFFFF'  # Background color for the legend
})
```

We also define the physical parameters of the system:

```python
# Physical parameters
Ω = 1  # Rabi frequency 
Δ = 1  # Detuning
```

The selection of the $\Omega$ and $\Delta$ values, in this case, is arbitrary and we consider them equal to $1$ for practical purposes. Now, we define the time array.

```python
# Time parameters
t_0 = 0                        # initial time (s)
t_f = 10                       # final time   (s)
Δt = 0.01                      # step size    (s)
n = int((t_f - t_0) / Δt) + 1  # iterations
t = np.linspace(t_0, t_f, n)   # time vector  (s)
```

To solve this system of PDE, we will use fourth-order Runge-Kutta method. We define the function in $\textcolor{yellow}{\texttt{Python}}$ as follows:

```python
# Fourth-Order Runge-Kutta function
def RK4(f, x0, y0, z0, h):
    k1y = h * f(x0, y0, z0)[0]
    k1z = h * f(x0, y0, z0)[1]
    #
    k2y = h * f(x0 + h / 2, y0 + k1y / 2, z0 + k1z / 2)[0]
    k2z = h * f(x0 + h / 2, y0 + k1y / 2, z0 + k1z / 2)[1]
    #
    k3y = h * f(x0 + h / 2, y0 + k2y / 2, z0 + k2z / 2)[0]
    k3z = h * f(x0 + h / 2, y0 + k2y / 2, z0 + k2z / 2)[1]
    #
    k4y = h * f(x0 + h, y0 + k3y, z0 + k3z)[0]
    k4z = h * f(x0 + h, y0 + k3y, z0 + k3z)[1]

    # Approximation
    y1 = y0 + (k1y + 2 * k2y + 2 * k3y + k4y) / 6
    z1 = z0 + (k1z + 2 * k2z + 2 * k3z + k4z) / 6
    return y1, z1
```

This function takes the following inputs:
- $f$: functions to evaluate.
- $x_{0}$: initial condition (independent variable).
- $y_{0}$: initial condition (dependent variable).
- $z_{0}$: initial condition (dependent variable).
- $h$: step size.

The next step is to define the function that contains the differential equations:

```python
# Differential Equations
def f(t, c1, c2):
    return complex(-1j * Ω * np.exp(1j * Δ * t) * c2), complex(-1j * Ω * np.exp(-1j * Δ * t) * c1)
```

We initialize our storage arrays:

```python
# Initial arrays
c1, c1[0] = np.zeros(n, dtype = 'complex'), 1 + 0j
c2, c2[0] = np.zeros(n, dtype = 'complex'), 0 + 0j
```

Finally, we perform the $\texttt{for}$ cycle:

```python
# RK4 method evaluation
for i in range(n - 1):
    c1[i + 1], c2[i + 1] = RK4(f, t[i], c1[i], c2[i], Δt)
```

And finally, we plot our results:

```python
# Plot
plt.figure(figsize = (10, 6))
plt.title(r'Rabi oscillations in an optically-driven two-level atom')
plt.xlabel(r'$t(s)$')
plt.ylabel(r'Population')

# 
plt.plot(t, np.abs(c1) ** 2, label = r'$|\tilde{c}_{1}(t)|^2$', color = '#D40000')
plt.plot(t, np.abs(c2) ** 2, label = r'$|\tilde{c}_{2}(t)|^2$', color = '#0000FF')
plt.plot(t, np.abs(c1) ** 2 + np.abs(c2) ** 2, label = r'$|\tilde{c}_{1}(t)|^2 + |\tilde{c}_{1}(t)|^2$', color = '#0B790B')

# Final parameters
plt.legend(loc = 'lower right', borderpad = 0.2)
plt.tight_layout()
```
